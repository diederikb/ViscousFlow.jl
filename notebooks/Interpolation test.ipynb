```julia
ddf = DDF(ddftype=Fields.M4prime)
x=range(-3, stop=2, length=1000)
x_left=range(-4, stop=-2, length=1000)
x_right=range(2, stop=4, length=1000)
plot(x,ddf.(x),legend=false,size=[150,100],linecolor=:red,yticks = 0:0.5:1,linewidth=3)
plot!(x_left,ddf.(x_left),legend=false,linecolor=:black,yticks = 0:0.5:1,linewidth=3)
plot!(x_right,ddf.(x_right),legend=false,linecolor=:black,yticks = 0:0.5:1,linewidth=3)
# savefig("M4prime.pdf")
```

```julia
function biot_savart(xg,xp,yp,Γ,dir)
    x = xg[1] - xp
    y = xg[2] - yp
    θ = atan(y,x)
    r = sqrt(x^2+y^2)
    if dir == 1
        return -Γ/(2*π*r)*sin(θ)
    else
        return Γ/(2*π*r)*cos(θ)
    end
end
```

### Biot-Savart velocities at edges

```julia
# primal u edges
xg,yg = coordinates(q_primal.u,dx=dx)
coord = [(x,y) for x in xg, y in yg]
for i in 1:size(coord,1), j in 1:size(coord,2)
    q_primal.u[i,j] = biot_savart(coord[i,j],xp.u[1],xp.v[1],1,1)
end

# primal v edges
xg,yg = coordinates(q_primal.v,dx=dx)
coord = [(x,y) for x in xg, y in yg]
for i in 1:size(coord,1), j in 1:size(coord,2)
    q_primal.v[i,j] = biot_savart(coord[i,j],xp.u[1],xp.v[1],1,2)
end  

# dual nodes u
xg,yg = coordinates(w_dual,dx=dx)
coord = [(x,y) for x in xg, y in yg]
for i in 1:size(coord,1), j in 1:size(coord,2)
    w_dual[i,j] = biot_savart(coord[i,j],xp.u[1],xp.v[1],1,2)
end 
```
